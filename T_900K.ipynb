```python
import numpy as np
import wgsceria as m
```

```python
# Boundary conditions
T = 900    # K
PCO  = 0.34  # bar
PH2O = 0.64  # bar
PCO2 = 0.01  # bar
PH2  = 0.01  # bar
```

```python
K, kf, kr = m.get_k(T, [PCO, PH2O, PCO2, PH2])

# Seed
theta0 = np.zeros(13)
theta0[11] = 1.0   # empty Ce
theta0[12] = 1.0   # empty O

# Integration until convergence
theta = m.int_conv(kf, kr, theta0)

# Get rates
rates = m.get_rates(theta, kf, kr)
```

```python
# Calculate degrees of rate control (DRC)
all_drc = np.array([m.get_drc(K, kf, theta, r) for r in range(15)])
```

```python
for r in range(15):
    print(
        f"Step {r+1:2d} | "
        f"dE = {m.dE[r]:6.2f} eV | "
        f"Ea = {m.Ea[r]:6.2f} eV | "
        f"rate = {rates[r]: .2e} | "
        f"DRC = {all_drc[r]: .4f}"
    )

table = np.column_stack([np.arange(len(m.dE))+1, m.dE, m.Ea, rates, all_drc])
table_header = "Step\tΔE (eV)\tEa (eV)\tr (s^-1)\tDRC"
np.savetxt('T_900K.txt', table, fmt=["%d", "%.2f", "%.2f", "%.2e", "%.2f"], delimiter="\t", header=table_header)
```
